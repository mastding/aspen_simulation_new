```python
import win32com.client as win32
aspen = win32.gencache.EnsureDispatch("Apwn.Document")
aspen.InitFromArchive2(r"D:\aspen\bkpfile\output_20251226_151605.bkp")
aspen.Visible = True
```

```python
def dfs(node, prefix=""):
    lines = []
    if node.Dimension == 0:
        line = f"{prefix}\\{node.Name} : {node.Value}"
        lines.append(line)
    else:
        line = f"{prefix}\\{node.Name}"
        lines.append(line)
        for child in node.Elements:
            lines.extend(dfs(child, prefix=f"{prefix}\\{node.Name}"))
    return lines
```

添加组分前

```python
node1 = aspen.Tree.FindNode(r"\Data\Streams\1\Input")
output1 = dfs(node1)
```

添加后

```python
node2 = aspen.Tree.FindNode(r"\Data\Streams\1\Input")
output2 = dfs(node2)
```

```python
# 比较输出
diff_lines = []
for i, (line1, line2) in enumerate(zip(output1, output2)):
    if line1 != line2:
        diff_lines.append(f"行 {i+1}:\n  - {line1}\n  + {line2}")

# 输出比较结果
if diff_lines:
    print("发现差异:")
    for diff in diff_lines:
        print(diff)
else:
    print("输出完全相同")
```
